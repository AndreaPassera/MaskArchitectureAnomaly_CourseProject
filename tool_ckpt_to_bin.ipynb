```python
import torch

print("Starting conversion of the model from .ckpt to .bin...")

# Paths
ckpt_path = "/content/drive/MyDrive/FAIMDL Project/checkpoints/step5_head_only_lightning/best-miou-epoch=09.ckpt"
bin_path = "/content/drive/MyDrive/FAIMDL Project/eomt_coco_finetuned.bin"

# 1. Load model
_original_torch_load = torch.load
torch.load = lambda *args, **kwargs: _original_torch_load(*args, **{**kwargs, 'weights_only': False})
checkpoint = torch.load(ckpt_path, map_location='cpu')
torch.load = _original_torch_load

# 2. Extract just the weights
pure_weights = checkpoint['state_dict']

# 3. Save as .bin file extension
torch.save(pure_weights, bin_path)

print(f"Conversion copleated!")
print(f"Model is now saved in: {bin_path}")
```
